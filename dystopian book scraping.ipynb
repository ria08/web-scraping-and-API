```python
page = 1
names = []
total_ratings = []
avg_rating = []
author=[]
score=[]
total_votes=[]
while page != 41:
    url = f"https://www.goodreads.com/list/show/47.Best_Dystopian_and_Post_Apocalyptic_Fiction?page={page}"
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36',"Accept": 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7', 'sec-ch-ua': 'Chromium";v="130", "Google Chrome";v="130", "Not?A_Brand";v="99', 'sec-ch-ua-mobile':'?0', 'sec-ch-ua-platform': 'Windows', 'sec-fetch-dest': 'document', 'sec-fetch-mode': 'navigate', 'sec-fetch-site': 'same-origin', 'sec-fetch-user': '?1', 'upgrade-insecure-requests': '1'}
    webpage=requests.get(url, headers=headers).text
    soup =BeautifulSoup(webpage, 'lxml')
    dyst_fiction=soup.find_all('tr', itemtype="http://schema.org/Book")
    for book in dyst_fiction:
        if book.find('td', width= '100%') is not None:
            
            name = book.find('a',class_="bookTitle").text.strip()
            names.append(name)
            
            authors = book.find('a',class_="authorName").text.strip()
            author.append(authors)
            
            rating_text = book.find('span', class_="minirating").text.strip()
            
            scoring = book.find('span',class_="minirating").text.strip().split()
            ascores=scoring[0]
            avg_rating.append(ascores)
            
            rates = scoring[4]
            total_ratings.append(rates)
            
            voted= book.find('span',class_="smallText uitext").text.strip().split()        
            scores=voted[1]
            score.append(scores)
            
            vote=voted[3]
            total_votes.append(vote)
            
        

    page += 1
```

```python
df = pd.DataFrame({'book_title': names,
                   'author': author,
'ratings': total_ratings,
'avg_rating': avg_rating,

'score': score,
 'total_votes':total_votes                   
})
df
```
